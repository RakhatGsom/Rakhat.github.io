```python
%%sql $CONNECT_DATA
    SELECT * FROM airports;
```

```python
%%sql $CONNECT_DATA
SELECT airport_name, airport_code, city FROM airports;
```

```python
%%sql $CONNECT_DATA
Select distinct aircrafts.model FROM airports
Join flights on airports.airport_code=flights.departure_airport
Join aircrafts on flights.aircraft_code=aircrafts.aircraft_code
WHERE airports.city IN ('Yakutsk', 'Abakan');
```

```python
%%sql $CONNECT_DATA
Select distinct aircrafts.model FROM aircrafts
Left join flights on aircrafts.aircraft_code=flights.aircraft_code
Left join airports on flights.departure_airport=airports.airport_code
Where airports.city IS NULL OR airports.city != 'Yakutsk';
```

```python
%%sql $CONNECT_DATA
SELECT departure_city.airport_name as departure_city,arrival_city.airport_name as arrival_city
FROM flights f
JOIN ticket_flights tf on f.flight_id = tf.flight_id
JOIN tickets t on tf.ticket_no = t.ticket_no
JOIN airports as departure_city on f.departure_airport = departure_city.airport_code
JOIN airports as arrival_city on f.arrival_airport = arrival_city.airport_code
WHERE t.passenger_name = 'ELLA DMITRIEVA';
```
