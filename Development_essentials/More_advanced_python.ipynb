```python
def Rakhat_finds_missing_numbers(sequence):
    missing_numbers = []

    # Iterate through the sequence to find missing numbers
    for i in range(len(sequence) - 1):
        diff = sequence[i + 1] - sequence[i]
        if diff > 1:
            missing_numbers.extend(range(sequence[i] + 1, sequence[i + 1]))

    return missing_numbers

# Example that was given for the final test
input_sequence1 = [-1, 2, 3, 4, 6, 7, 10]
result1 = Rakhat_finds_missing_numbers(input_sequence1)
print(result1)

input_sequence2 = [5, 6, 8, 10, 11, 12, 15]
result2 = Rakhat_finds_missing_numbers(input_sequence2)
print(result2)

input_sequence3 = [504, 505, 510]
result3 = Rakhat_finds_missing_numbers(input_sequence3)
print(result3)
```
