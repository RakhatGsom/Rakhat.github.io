"Рассчитайте наценку на ювелирные изделия по данным на листе ""Задание 2"". Соблюсти следующие условия:
1. Период анализа - с 1.04.23 по 10.04.23
2. В анализе должны участвовать только продажи изделий с общей суммой и себестоимостью больше 0 рублей
3. В ответе необходимо получить два значения наценки - среднюю наценку на ювелирные изделия из золота и среднюю наценку на ювелирные изделия из серебра. Сгруппировать изделия по металлу нужно на основе поля ""Товарная группа"". Товарная группа изделий из серебра начинается на ""СИ"", все остальные изделия - это золото."


```python
# Фильтрация данных по указанному периоду
start_date = '2023-04-01'
end_date = '2023-04-10'
df['Датапродажи'] = pd.to_datetime(df['ДатаПродажи'])
filtered_df = df[(df['Датапродажи'] >= start_date) & (df['Датапродажи'] <= end_date)]

# Фильтрация по условиям
filtered_df = filtered_df[(filtered_df['ОбщаяСумма'] > 0) & (filtered_df['Себестоимость'] > 0)]

# Группировка по металлу (золото/серебро)
gold_group = filtered_df[~filtered_df['ТоварнаяГруппа'].str.startswith('СИ')]
silver_group = filtered_df[filtered_df['ТоварнаяГруппа'].str.startswith('СИ')]

# Расчет средней наценки
avg_markup_gold = (gold_group['ОбщаяСумма'] - gold_group['Себестоимость']).mean()
avg_markup_silver = (silver_group['ОбщаяСумма'] - silver_group['Себестоимость']).mean()

print(f"Средняя наценка на ювелирные изделия из золота: {avg_markup_gold}")
print(f"Средняя наценка на ювелирные изделия из серебра: {avg_markup_silver}")
```
